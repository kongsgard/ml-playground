```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['mathtext.fontset'] = 'cm'
```

```python
fig, axs = plt.subplots(4, 2, figsize=(16,16))

z = np.arange(-10, 10, .1)
zero = np.zeros(len(z))

# Sigmoid
axs[0, 0].axis('off')

annotation = r'Sigmoid' + '\n\n' + r'$\phi(z) = \frac{1}{1 + e^{-z}}$'
axs[0, 0].text(0.5, 0.5, annotation, horizontalalignment='center', verticalalignment='center', fontsize=30)

y = 1 / (1 + np.exp(-z))
axs[0, 1].plot(z, y)

# Hyberbolic Tangent (Tanh)
axs[1, 0].axis('off')

annotation = r'Hyperbolic Tangent (Tanh)' + '\n\n' + r'$\phi(z) = \frac{e^z - e^{-z}}{e^z + e^{-z}}$'
axs[1, 0].text(0.5, 0.5, annotation, horizontalalignment='center', verticalalignment='center', fontsize=30)

y = np.tanh(z)
axs[1, 1].plot(z, y)

# Rectified Linear Unit (ReLU)
axs[2, 0].axis('off')

annotation = r'Rectified Linear Unit (ReLU)' + '\n\n' + r'$\phi(z) = \max(0, z)$'
axs[2, 0].text(0.5, 0.5, annotation, horizontalalignment='center', verticalalignment='center', fontsize=30)

y = np.max([zero, z], axis=0)
axs[2, 1].plot(z, y)

# Leaky ReLU
axs[3, 0].axis('off')

annotation = r'Leaky ReLU' + '\n\n' + r'$\phi(z) = \max(x, ax), \: a \leq 1$'
axs[3, 0].text(0.5, 0.5, annotation, horizontalalignment='center', verticalalignment='center', fontsize=30)

y = np.max([z, 0.1*z], axis=0)
axs[3, 1].plot(z, y)

plt.savefig('activation_functions.eps', format='eps', dpi=1000)
```
